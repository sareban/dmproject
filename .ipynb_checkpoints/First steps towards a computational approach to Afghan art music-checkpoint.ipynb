### A few words on the dataset

The data consists in Afghan classical themes gathered by Prof. John Baily, Goldsmith University, and compiled into a document prepared at the occasion of the 2014 Winter School of the Afghan National Institute of Music, Kabul (Afghanistan). The music is notated following the common, afghanized version of the Hindustani Saregam notation. Because this style of music is based on variations (both melodic and rythmic) on themes, and to some extent improvisations within the modes (known as ragas, individually raags), compositions are fairly short in terms of notes, usually a few cycles per movement of a composition. 

The dataset that was provided to us consists of 43 compositions of various length and rythmic cycles, in 15 different modes.

### Importing the data

Although we have prepared several level segmentations of our data, we focus here on an approach that considers compositions in their totality; we do not discriminate between movements of said compositions, which are mostly divided between a part called "asthai", the main theme, and "antara", when the melody usually goes to the upper part of the range. 

We begin by creating a list of the symbols. It goes from Nii (low B) up to Ma1 (high F). v stands for either a silence or a continuation of the preceding note, usually for the musician to decide.

Sa is the tonic, Pe the fifth. If we were to speak in terms of western staff notation, we would use C as the tonic, but that is simply a convention of transcription while these modes may be played from a different tonic.

### Pitch distribution

Compositions are imported and ready for manipulation. Next, we proceed to count pitches in each composition. We then compute their proportions within the compositions, and these are the first features we explore.

```python
def count_notes(n):
    count = list()
    for note in symbols :
        count.append(n.count(note));
    total = sum(count)
    count = [(x*100/total) for x in count]
    count.append(total)
    return count
```

```python
total_count = list()
for composition in df_full.Composition :
    total_count.append(count_notes(composition));
df_full_count = pd.DataFrame(total_count, columns=symbols+["Total"])
```

```python
df_full_count.head(5)
```

Obviously, with this method of counting, we have a redundancy issue: notes out of the main octave are counted twice, eg: Sa1 counts for Sa1 as well as for Sa. This can easily be fixed without changing our straightforward counting method. Regexp parsing could also have been used to avoid this problem.

```python
df_full_count.Ni = df_full_count.Ni - df_full_count.Nii
df_full_count.Sa = df_full_count.Sa - df_full_count.Sa1
df_full_count.Ra = df_full_count.Ra - df_full_count.Ra1
df_full_count.Re = df_full_count.Re - df_full_count.Re1
df_full_count.Ga = df_full_count.Ga - df_full_count.Ga1
df_full_count.Ge = df_full_count.Ge - df_full_count.Ge1
df_full_count.Ma = df_full_count.Ma - df_full_count.Ma1
df_full_count.Total = df_full_count.Total - df_full_count.Nii -df_full_count.Sa1-df_full_count.Ra1-df_full_count.Re1-df_full_count.Ga1-df_full_count.Ge1-df_full_count.Ma1
```

A prior plot of pitch distribution by composition shows a few outliers but a rather consistent trend among pieces.

```python
df_full_count.iloc[:,0:20].plot(figsize=(10,10))
plt.title("Pitch distribution by composition")
plt.xlabel('Composition no')
plt.ylabel('%')
plt.show()
```

Still, the imbalance between modes representation in the dataset is not leveled, for such we should aggregate the numbers by mode.

```python
count_raag = list()

for raag in df_full.Raag.unique() :
    count = [0] * 21
    for index in df_full.loc[df_full['Raag'] == raag].index.values.tolist() :
        count = count + df_full_count.iloc[index-1];
    total = count[20]
    count = [(x/len(df_full.loc[df_full['Raag'] == raag].index.values.tolist())) for x in count[0:20]]
    count.append(total)
    count.append(raag)
    count_raag.append(count);
    
df_count_raag = pd.DataFrame(count_raag, columns=symbols+["Total","Raag"])
df_count_raag.set_index('Raag', inplace=True)
```

```python
df_count_raag.head(5)
```

### Notes bigrams
the following is adapated from the tutorial exercise to fit our data, and try to get more insight about it.

```python
sequences = [ [interval.strip() for interval in re.sub( r"([A-Z])", r" \1", row.Composition).replace("v", "").split() ]  for i, row in df_full.iterrows()]
```

```python
from collections import Counter

all_bigrams = []

for sequence in sequences:
    bigrams = list(zip(*[sequence[i:] for i in range(2)]))
    bigram_counts = Counter(bigrams)
    all_bigrams.append(bigram_counts)
```

```python
total_counts = sum(all_bigrams, Counter())
```

```python
total_counts.most_common(20)
```

```python
int1s = [ int1 for (int1, int2), count in total_counts.items() ]
int2s = [ int2 for (int1, int2), count in total_counts.items() ]
counts = [ count for (int1, int2), count in total_counts.items() ]
```

```python
transitions = pd.DataFrame( {
    'note1':int1s,
    'note2':int2s,
    'counts':counts})
```

```python
unique_notes = set(list(transitions.note1.unique())+list(transitions.note2.unique()))
```

```python
note_class = {
    "Nii" : -1,
     "Sa" : 0,
     "Ra" : 1,
     "Re" : 2,
     "Ga" : 3,
     "Ge" : 4,
     "Ma" : 5,
     "Me" : 6,
     "Pe" : 7,
     "Da" : 8,
     "De" : 9,
     "Na" : 10,
     "Ni" : 11,
     "Sa1": 0,
     "Ra1": 1,
     "Re1": 2,
     "Ga1": 3,
     "Ge1": 4,
     "Ma1": 5}
```

```python
transitions['pc1'] = transitions['note1'].map(note_class)
transitions['pc2'] = transitions['note2'].map(note_class)

transitions.head()
#transitions.pc2.isnull().values.any()
```

```python
df = pd.DataFrame(index=np.arange(-1,12), columns=np.arange(-1,12)).fillna(0)
```

```python
for index, row in transitions.iterrows():
    int1 = row['pc1']
    int2 = row['pc2']
    df.loc[int1,int2] += row['counts']
```

### Transition matrix (Markov table as a heatmap)

### Conditional distributions of interval successions

```python
freqs = df.div(df.sum(axis=1), axis=0)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(freqs, cmap='coolwarm')
plt.show()
```

### Entropy

We compute the entropy for each pitch class based on the full aggregated dataset, in order to get an idea of global entropy levels of the music considered.

```python
fig, axes = plt.subplots(nrows=12, ncols=1,figsize=(10,50))

entropies = []

for ax, i in zip(axes, range(-1,12)): #[-1,11]
    marginal = freqs.loc[i,:]
    e = round(scipy.stats.entropy(marginal),3)
    entropies.append(e)
    ax.bar(np.arange(-1,12), marginal )
    ax.set_title(f'Pitch Class: {i}, Entropy: {e}')
    ax.set_xticks(np.arange(-1,12))

plt.tight_layout()
plt.show();
```

### Comparing means
We then compare it to the entropies computed for the exercise dataset. We do a T-Test to verify whether there is a significant difference between the two datasets.

```python
entropies_exercise = [1.628, 2.244, 1.89, 2.132, 2.205, 1.742, 2.177, 2.193, 2.006, 2.142, 1.939, 1.627]
```

```python
ttest_ind(entropies, entropies_exercise)
```

Indeed, with a t-statistic of 3.3 and a pvalue o 0.003, we may reject the null hypothesis at 0.01, and claim that the two groups are significantly different, meaning that entropies are found to be lower for our Afghan music dataset

### Computing Markov tables and entropies for single ragas
Next we assemble the previous code into function that may be used on single modes of our dataset, allowing us for instance to compare the resulting entropies.

```python
def process_compositions(df_full, note_class):
    sequences = [ [interval.strip() for interval in re.sub( r"([A-Z])", r" \1", row.Composition).replace("v", "").split() ]  for i, row in df_full.iterrows()]
    all_bigrams = []

    for sequence in sequences:
        bigrams = list(zip(*[sequence[i:] for i in range(2)]))
        bigram_counts = Counter(bigrams)
        all_bigrams.append(bigram_counts)
        
    total_counts = sum(all_bigrams, Counter())    
    int1s = [ int1 for (int1, int2), count in total_counts.items() ]
    int2s = [ int2 for (int1, int2), count in total_counts.items() ]
    counts = [ count for (int1, int2), count in total_counts.items() ]
     
    transitions = pd.DataFrame( {
        'note1':int1s,
        'note2':int2s,
        'counts':counts,
        } )
    
    transitions['pc1'] = transitions['note1'].map(note_class)
    transitions['pc2'] = transitions['note2'].map(note_class)
    
    df = pd.DataFrame(index=np.arange(-1,12), columns=np.arange(-1,12)).fillna(0)
    for index, row in transitions.iterrows():
        int1 = row['pc1']
        int2 = row['pc2']
        df.loc[int1,int2] += row['counts']
    return df
```

```python
def get_entropies(df_full, note_class):
    
    df = process_compositions(df_full, note_class)
    freqs = df.div(df.sum(axis=1), axis=0)
    entropies = []

    for ax, i in zip(axes, range(-1,11)): #[-1,11]
        marginal = freqs.loc[i,:]
        e = round(scipy.stats.entropy(marginal),3)
        entropies.append(e)
    
    return entropies
```

```python
def print_markov(df_full, note_class):
    
    df = process_compositions(df_full, note_class)
    plt.figure(figsize=(10,10))
    sns.heatmap(df, cmap='coolwarm')
    plt.xlabel('Second Note')
    plt.ylabel('First Note')
    plt.show()
```

```python
for raag in df_full.Raag.unique() :
    plt.bar(range(-1,11), get_entropies(df_full.loc[df_full['Raag'] == raag], note_class))
    plt.title(raag)
    plt.xlabel('Pitch class')
    plt.ylabel('Entropy')
    plt.show()
```

```python
for raag in df_full.Raag.unique() :
    plt.bar(range(-1,12), print_markov(df_full.loc[df_full['Raag'] == raag], note_class))
    plt.title(raag)
    plt.xlabel('Pitch class')
    plt.ylabel('Entropy')
    plt.show()
```
